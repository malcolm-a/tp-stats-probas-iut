### 1.3 Liaison linéaire

```python
def mean(dat: list):
    return sum(dat)/len(dat)
```

```python
def quantile(dat: list, p: float):
    sdat = sorted(dat)
    i = round(p*(len(sdat)))
    return sdat[i]
```

```python
def median(dat: list):
    return quantile(dat, 0.5)
```

```python
from math import sqrt

def stdev(dat: list):
    st_sum = 0
    for i in range(len(dat)):
        st_sum += abs(mean(dat) - dat[i])**2
    return sqrt(st_sum/len(dat))
```

```python
def cofvar(dat: list):
    return stdev(dat) / mean(dat)
```

```python
def valrange(dat: list):
    return max(dat) - min(dat)
```

```python
def iquarange(dat: list):
    return quantile(dat, 0.75) - quantile(dat, 0.25)
```

```python
def scale(data: list, center=True, scale=True):
    new_data = data.copy()

    if center:
        data_mean = mean(data)
        for i in range(len(new_data)):
            new_data[i] -= data_mean
            
    if scale:
        data_stdev = stdev(data)
        for i in range(len(new_data)):
            new_data[i] /= data_stdev
    return new_data
```

```python
def coefcor(x: list, y: list):
    xc = scale(x, scale=False)
    yc = scale(y, scale=False)

    z = [i for i in xc]
    for i in range(len(z)):
        z[i] *= yc[i]
    
    z_bp = mean(z) / (stdev(x) * stdev(y))
    
    return z_bp

coefcor([1,2,3,4],[2,4,6,8])
```

### 1.4 Liaison non-linéaire

```python
def to_ranks(data):
    sorted_data = sorted(data)
    return [sorted_data.index(x) + 1 for x in data]
```

```python
def coefspearman(x: list, y: list):
    return coefcor(to_ranks(x), to_ranks(y))
```

```python
def coefspearman2(x, y):
    return 1 - (6 * sum((abs(xi - yi) ** 2) for xi, yi in zip(to_ranks(x), to_ranks(y)))) / (len(x) ** 3 - len(x))
```

```python
def split_by(x, y):
    d = dict()
    for yi in set(y):
        d[yi] = []
    for i in range(len(x)):
        d[y[i]].append(x[i])
    return dict(sorted(d.items()))
```

```python
def describe_by(x, y, width=5):
    d = split_by(x, y)
    
    # Affichage de la première ligne
    print("                 ", end='')
    for key in d.keys():
        print(f"{key:>{width}}", end='')
        
    # Affichage des moyennes
    print("\nMoyenne         :", end='')
    for k, v in d.items():
        print(f"{round(mean(v), 2):>{width}}", end='')
        
    # Affichage des médianes
    print("\nMédiane         :", end='')
    for k, v in d.items():
        print(f"{round(median(v), 2):>{width}}", end='')

    # Affichage des écarts-types
    print("\nÉcart-type      :", end='')
    for k, v in d.items():
        print(f"{round(stdev(v), 2):>{width}}", end='')

    # Affichage des coefficients de variation
    print("\nCoef. de var.   :", end='')
    for k, v in d.items():
        print(f"{round(cofvar(v), 2):>{width}}", end='')

    # Affichage des étendues
    print("\nÉtendue         :", end='')
    for k, v in d.items():
        print(f"{round(valrange(v), 2):>{width}}", end='')

    # Affichage des écarts interquartiles
    print("\nÉc. interquart. :", end='')
    for k, v in d.items():
        print(f"{round(iquarange(v), 2):>{width}}", end='')

    # Affichage des premiers quartiles
    print("\nPrem.  quartile :", end='')
    for k, v in d.items():
        print(f"{round(quantile(v, 0.25), 2):>{width}}", end='')

    # Affichage des troisièmes quartiles
    print("\nTrois. quartile :", end='')
    for k, v in d.items():
        print(f"{round(quantile(v, 0.75), 2):>{width}}", end='')
```

```python
def contingency_table(var1: list, var2: list, freq=False, sort_keys=True, reverse=False):
    d = dict()
    for v1 in set(var1):
        for v2 in set(var2):
            d[(v1, v2)] = 0
    for i in range(len(var1)):
        d[(var1[i], var2[i])] += 1
    if freq:
        d_sum = sum(d.values())
        for k, v in d.items():
            d[k] = round(d[k] / d_sum, 2)
    
    
    # return sorted
    if sort_keys:
        return dict(sorted(d.items(), reverse=reverse))
    else:
        return dict(sorted(d.items(), key=lambda item: item[1], reverse=reverse))
```

```python
def margin(d: dict, m=0):
    if m == 1:
        # initialisation du nouveau dictionnaire
        rows = dict()
        for key in d.keys():
            rows[key[0]] = 0
        # remplissage du nouveau dictionnaire
        for key, val in d.items():
            rows[key[0]] = round(rows[key[0]] + val, 2)
        return rows
    elif m == 2:
        # initialisation du nouveau dictionnaire
        cols = dict()
        for key in d.keys():
            cols[key[1]] = 0
        # remplissage du nouveau dictionnaire
        for key, val in d.items():
            cols[key[1]] = round(rows[key[1]] + val, 2)
        return cols
    else:
        return sum(d.values())
```

```python
def get_gdistr(data: list):
    return {e: (list(data.values()).count(e), list(data.values()).count(e)/len(data)) for e in list(set(data))}
```

```python
def dchi2(cont):
    marg1 = margin(cont, 1)
    marg2 = margin(cont, 2)
    N = margin(cont, 0)
    d2 = 0
    for m1 in marg1.keys():
        for m2 in marg2.keys():
            n_the = (marg1[m1] * marg2[m2]) / N
            n_obs = cont[(m1,m2)]
            d2 += ((n_obs - n_the)**2) / n_the
    return d2
```

```python
print(dchi2(contingency_table(var1, var2)))
```

```python
from math import sqrt

def v_cramer(var1: list, var2: list):
    ct = contingency_table(var1, var2)
    return sqrt(dchi2(ct) / (len(var1) * (min(len(set(var1)), len(set(var2))) -1)))
```

```python
def scale(data: list, center=True, scale=True):
    new_data = data.copy()

    if center:
        data_mean = mean(data)
        for i in range(len(new_data)):
            new_data[i] -= data_mean
            
    if scale:
        data_stdev = stdev(data)
        for i in range(len(new_data)):
            new_data[i] /= data_stdev
    
    return new_data
```

```python
# Tracé du diagramme de dispersion
plt.scatter(margarine, divorce)
# Paramètres graphiques
plt.title('divorces et margarine dans le Maine') 
plt.xlabel('consommation de margarine')  
plt.ylabel('nb de divorces pour 1000 habitants')                    
plt.axis('equal')                 
plt.grid()
# Régression linéaire (non-demandé, fonctions numpy trouvées grâce à chat-gpt)
coefficients = np.polyfit(margarine, divorce, 1)
pente = coefficients[0]
interception = coefficients[1]
plt.plot(np.array(margarine), pente * np.array(margarine) + interception, color='red', label='Régression linéaire')
# Affichage
plt.show()
```

```python
# Calcul de R^2
predicted = pente * np.array(margarine) + interception
r_squared = 1 - (np.sum((divorce - predicted) ** 2) \
                / np.sum((divorce - np.mean(divorce)) ** 2))

print("Coefficient de détermination (R^2) :", r_squared)
```

On observe une corrélation très marquée entre la consommation de margarines et le taux de divorce. Ceci étant, bien qu'on observe une *corrélation*, il est difficile de pouvoir parler de *causalité*. Un lien peut exister, mais c'est probablement une coïncidence.
